## Solving the integer LPP using python

```python
# Create the model
model = LpProblem(name="our-problem", sense=LpMaximize)
```

```python
# Initialize the decision variables
x = LpVariable(name="x", lowBound=0)
y = LpVariable(name="y", lowBound=0)
```

```python
# Add the constraints to the model
model += (2 * x + y <= 20, "red_constraint")
model += (4 * x - 5 * y >= -10, "blue_constraint")
model += (-x + 2 * y >= -2, "yellow_constraint")
model += (-x + 5 * y == 15, "green_constraint")
```

```python
# Add the objective function to the model
model += x + 2 * y
```

```python
model
```

```python
# Solve the problem
status = model.solve()

#.solve() calls the underlying solver, modifies the model object,
#and returns the integer status of the solution, which will be 1 if the optimum is found.
#https://www.coin-or.org/PuLP/constants.html#pulp.constants.LpStatus
```

```python
print(f"status: {model.status}, {LpStatus[model.status]}")
```

```python
print(f"objective: {model.objective.value()}")
```

```python
for var in model.variables():
    print(f"{var.name}: {var.value()}")
```

```python
 for name, constraint in model.constraints.items():
    print(f"{name}: {constraint.value()}")
```

```python
model.solver
#The output informs that the solver is CBC. PuLP called the default one.
# Can be changed to GLPK etc
```

# Mixed Integer Programming Case

### Adding constraint for the decision variables to be integers

```python
## new feasible region is a set of discrete points
Image("FS_integer.png")
```

### To define an integer or binary variable,  just pass cat="Integer" or cat="Binary" to LpVariable. Everything else remains the same:

```python
# Create the model
model = LpProblem(name="small-problem", sense=LpMaximize)

# Initialize the decision variables: x is integer, y is continuous
x = LpVariable(name="x", lowBound=0, cat="Integer")
y = LpVariable(name="y", lowBound=0)

# Add the constraints to the model
model += (2 * x + y <= 20, "red_constraint")
model += (4 * x - 5 * y >= -10, "blue_constraint")
model += (-x + 2 * y >= -2, "yellow_constraint")
model += (-x + 5 * y == 15, "green_constraint")

# Add the objective function to the model
model += lpSum([x, 2 * y])

# Solve the problem
status = model.solve()
```

```python
print(f"status: {model.status}, {LpStatus[model.status]}")

print(f"objective: {model.objective.value()}")

for var in model.variables():
    print(f"{var.name}: {var.value()}")
```
